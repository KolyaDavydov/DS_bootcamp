```python
import pandas as pd
import sqlite3
```

```python
try:
  conn = sqlite3.connect('../data/checking-logs.sqlite')
  print('Подключение состоялось')
except sqlite3.Error as error:
  print('Подключение не состоялось:', error)
```

## анализировать только пользователей, а не администраторов

```python
query = '''
SELECT *
FROM pageviews
WHERE uid LIKE 'user_%'
'''
pageview = pd.io.sql.read_sql(query, conn, parse_dates=['datetime'])
pageview
```

```python
pageview = pageview.groupby(by=pageview['datetime'].dt.date).agg({'uid': ['count']})
pageview = pageview.reset_index()
pageview.columns = ['date', 'view']
pageview
```

```python
query = '''
SELECT *
FROM checker
WHERE uid LIKE 'user_%'
'''
checker = pd.io.sql.read_sql(query, conn, parse_dates=['timestamp'])
checker
```

```python
checker = checker.groupby(by=checker['timestamp'].dt.date).agg({'uid': ['count']})
checker = checker.reset_index()
checker.columns = ['date', 'commits']
checker
```

## анализировать только те даты, когда были и просмотры, и коммиты `checker`

```python
df = pd.merge(left=pageview, right=checker, on='date', how='inner')
df
```

## используйте размер шрифта должен быть 8
## размер фигуры (15,8)

```python
plt = df.plot(x='date',
              y=['view', 'commits'],
              figsize=(15,8),
             fontsize=8,
             title='Views and commits per day',
             rot=90,
             style=['-.', '--'],
             color=['r','c'])
plt.get_lines()[0].set_marker('o')
plt.grid(axis='x')
```
