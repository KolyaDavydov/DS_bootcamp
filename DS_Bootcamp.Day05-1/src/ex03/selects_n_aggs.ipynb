```python
import pandas as pd
```

## Загрузите файл JSON, созданный в предыдущем упражнении, в фрейм данных.

```python
# установите CarNumber в качестве индексного столбца
df = pd.read_json('../data/auto.json',
                 orient='records')
df.set_index('CarNumber', inplace=True)
df
```

## Сделать следующие выборки

```python
# отобразить строки, где Fines больше 2100
df[df['Fines'] > 2100]
```

```python
# отображать только те строки, где Fines больше 2100 и Refund равен 2
df[(df['Fines'] > 2100) & (df['Refund'] == 2)]
```

```python
# отображать только те строки, где есть модели из списка:[’Focus’, ’Corolla’]
df[df['Model'].isin(['Focus', 'Corolla'])]
```

```python
# отображать только те строки, где номер автомобиля находится в списке:
# [’Y7689C197RUS’, ’92928M178RUS’, ’7788KT197RUS’, ’H115YO163RUS’, ’X758HY197RUS’]
nums = ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
df[df.index.isin(nums)]
```

## Cделать агрегаты с маркой и моделью

```python
# отображать медианные Fines, сгруппированные по маркам
df.groupby(by=['Make']).agg({'Fines': ['median']})
```

```python
# отображать медианные Fines, сгруппированные по Make и Model
df.groupby(by=['Make', 'Model']).agg({'Fines': ['median']})
```

```python
# отображать количество Fines, сгруппированных по Make и Model,
# чтобы понять, можно ли доверять медианным значениям
df.groupby(by=['Make', 'Model']).agg({'Fines': ['count']})
```

```python
# отображать минимальные и максимальные Fines, сгруппированные по Make и Model, чтобы лучше понять разницу
df.groupby(by=['Make', 'Model']).agg({'Fines': ['min', 'max']})
```

```python
# отображать стандартное отклонение Fines, сгруппированных по Make и Model, чтобы лучше понять отклонение
df.groupby(by=['Make', 'Model']).agg({'Fines': ['std']})
```

## Cделать агрегаты с номером автомобиля

```python
# отобразить номера автомобилей, сгруппированные по количеству штрафов в порядке убывания,
# мы хотим найти тех, кто чаще всего нарушал закон
df.groupby(by=['CarNumber']).agg({'Fines': ['count']}).sort_values(by=('Fines', 'count'),
                                                                  ascending=False)
```

```python
# выберите из исходного фрейма данных все строки,
# соответствующие номеру автомобиля top-1, мы хотим немного увеличить масштаб
# Примечание: я так понял нужно отобразить строки для авто с номером у которого больше всего штрафов
df[df.index == 'Y7689C197RUS']
```

```python
# отобразить номера автомобилей, сгруппированные по сумме штрафов в порядке убывания,
# мы хотим найти тех, кто заплатил больше всего
df.groupby(by=['CarNumber']).agg({'Fines': ['sum']}).sort_values(by=('Fines', 'sum'), ascending=False)
```

```python
# выберите из исходного фрейма данных все строки,
# соответствующие номеру автомобиля номер один, мы хотим немного увеличить масштаб
df[df.index == 'X758HY197RUS']
```

```python
# вывести таблицу, отвечающую на вопрос: есть ли номера автомобилей, которые были привязаны к разным моделям?
df.groupby(by=['CarNumber', 'Model']).agg({'Model': ['nunique']}).sort_values(by=('Model', 'nunique'), ascending=False)

# ВЫВОД: одинаковых номеров привязанных к разным маркам - НЕТ
```
