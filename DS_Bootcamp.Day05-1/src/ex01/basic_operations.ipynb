```python
import pandas as pd
```

## создайте представления данных с двумя столбцами: `datetime` и `user`, прочитав файл `feedviews.log`.

```python
df = pd.read_csv('../data/feed-views.log',
                names=['datetime', 'user'],
                parse_dates=['datetime'],
                sep='\t')

df.info()
```

```python
df['year'] = pd.DatetimeIndex(df['datetime']).year
df['month'] = pd.DatetimeIndex(df['datetime']).month
df['day'] = pd.DatetimeIndex(df['datetime']).day
df['hour'] = pd.DatetimeIndex(df['datetime']).hour
df['minute'] = pd.DatetimeIndex(df['datetime']).minute
df['second'] = pd.DatetimeIndex(df['datetime']).second

df
```

## Создание нового столбца 'Daytime'
- вам необходимо назначить конкретное значение времени суток, если час находится в определенном интервале, например, полдень, если час больше 11 и меньше или равен 17.
- 0 – 3.59 ночь, 4 – 6.59 раннее утро, 7 – 10.59 утро, 11 – 16.59 день, 17 – 19.59 ранний вечер, 20 – 23.59 вечер.
- используйте метод Cut для решения этой подзадачи
- назначить пользователя столбца в качестве индекса

```python
df['daytime'] = pd.cut(df['hour'],
                      bins=[-1, 3, 6, 10, 16, 19, 23],
                      labels=['night', 'early morning', 'morning', 'afternoon', 'early evening', 'evening'])
df.set_index('user', inplace=True)
df
```

## Вычислить количество элементов в каждой категории суток(daytime) в dataframe

```python
df['daytime'].value_counts()
```

## Отсортируйте значения в вашем фрейме данных по часам, минутам и секундам в порядке возрастания (одновременно, а не по одному)

```python
df.sort_values(by=['hour', 'minute', 'second'], inplace=True)
df
```

## Calculate the minimum and maximum for the hours and the mode for the daytime categories

```python
# Найдем максимальное значение в колонке "час" в категории "ночь"
max_night = df[df['daytime'] == 'night']
max_night = max_night['hour'].max()
max_night
```

```python
# Найдем минимальное значение в колонке "час" в категории "утро"
min_morning = df[df['daytime'] == 'morning']
min_morning = min_morning['hour'].min()
min_morning
```

```python
# Выведем этих товарищей
df[df['hour']==max_night]
```

```python
df[df['hour']==min_morning]
```

```python
# мода для hour
df['hour'].mode()
```

```python
# мода для daytime
df['daytime'].mode()
```
