## Обогатить фрейм данных, используя образец из этого фрейма данных

### создайте выборку из 200 новых наблюдений со random_state = 21
- в выборке не должно быть новых комбинаций номера, марки и модели автомобиля, чтобы весь набор данных был согласованным в этих терминах.
- ограничений по возврату и штрафам нет, вы можете взять любое значение из этих столбцов наугад и использовать его в отношении любого номера автомобиля

```python
sample = df.sample(n=200, random_state=21)
sample
```

```python
uniqueRefund = df['Refund'].unique()
uniqueRefund
```

```python
uniqueFines = df['Fines'].unique()
# uniqueFines
```

```python
sample['Refund'] = np.random.choice(uniqueRefund, size=len(sample))
sample['Fines'] = np.random.choice(uniqueFines, size=len(sample))
sample
```

### Объединить образец с исходным фреймом данных в новый фрейм данных concat_rows

```python
concat_rows = pd.concat(objs=[df, sample], ignore_index=True)
concat_rows
```

## Обогатить фрейм данных concat_rows новым столбцом с сгенерированными данными
- создайте серию с названием Year, используя случайные целые числа от 1980 до 2019 года.
- используйте np.random.seed(21) перед созданием Year
- объединить серию с dataframe и назвать ее 'fines'

```python
# Задаем параметры генерации случайных чисел
np.random.seed(21)

Year = pd.Series([np.random.randint(1980, 2019) for i in range(len(concat_rows))], name='Year')
Year
```

```python
fines = pd.concat([concat_rows, Year], axis='columns')
fines
```

## Обогатить фрейм данных данными из другого фрейма данных

### создать новый фрейм данных с номерами автомобилей и их владельцами
- получить самые популярные фамилии (файл `surname.json` можно найти во вложениях) в США
= создайте новую серию с фамилиями (они не должны иметь специальных символов, таких как запятые, скобки и т. д.) на основе собранных вами данных, количество должно быть равно количеству уникальных номеров автомобилей в выборке (используйте random_state = 21 )
- создайте владельцев данных с двумя столбцами: CarNumber и SURNAME.

```python
surname = pd.read_json('../data/surname.json', orient='values')
surname.columns = surname.iloc[0]
surname.drop(0, inplace=True)
surname
```

```python
car_numbers = concat_rows.drop_duplicates('CarNumber')['CarNumber']
car_numbers = car_numbers.to_frame(name='CarNumber').reset_index(drop=True)
car_numbers
```

```python
# Year = pd.Series([np.random.randint(1980, 2019) for i in range(len(concat_rows))], name='Year')
surnames = surname['NAME'].sample(n=len(car_numbers), random_state=21, replace=True)
surnames = surnames.to_frame(name='SURNAME').reset_index(drop=True)
surnames
```

```python
owners = pd.concat([car_numbers, surnames],axis=1, ignore_index=False)
owners
```

### удалите последние 20 наблюдений из фрейма данных `owners` и добавьте 3 новых наблюдения (они не совпадают с теми, которые вы добавляете в фрейм данных о штрафах)

## присоединитесь к обоим фреймам данных:


- новый фрейм данных должен содержать только номера автомобилей, которые существуют в обоих фреймах данных.


```python
fines_1 = pd.merge(left=fines, right=owners, on='CarNumber', how='inner')
fines_1
```

- новый фрейм данных должен содержать все номера автомобилей, существующие в обоих фреймах данных.


```python
fines_2 = pd.merge(left=fines, right=owners, on='CarNumber', how='outer')
fines_2
```

- новый фрейм данных должен содержать только номера автомобилей из фрейма данных о штрафах


```python
fines_3 = pd.merge(left=fines, right=owners, on='CarNumber', how='left')
fines_3
```

- новый фрейм данных должен содержать только номера автомобилей из фрейма данных владельцев.

```python
fines_4 = pd.merge(left=fines, right=owners, on='CarNumber', how='right')
fines_4
```

## создайте сводную таблицу из кадра данных о штрафах, она должна выглядеть так (значения представляют собой суммы штрафов), но со всеми годами (у вас значения могут быть другими):

```python
pd.pivot_table(fines,
               columns='Year',
               values='Fines',
               index=['Make', 'Model'],
               aggfunc={'Fines': np.sum})
```
